```python
import os
import sqlite3
import pandas as pd
import numpy as np

# Absolute paths setup
BASE_DIR = os.path.dirname(os.path.dirname(os.path.abspath('__file__')))
DB_PATH = os.path.join(BASE_DIR, 'bluestock_mf.db')
PROCESSED_DIR = os.path.join(BASE_DIR, 'data', 'processed')
os.makedirs(PROCESSED_DIR, exist_ok=True)

conn = sqlite3.connect(DB_PATH)
print(" Starting Day 4 Fund Performance Analytics Engine...")

# 1. Fetch tables from SQLite
df_nav = pd.read_sql_query("SELECT amfi_code, date, nav FROM fact_nav;", conn)
df_perf = pd.read_sql_query("SELECT amfi_code, return_3yr_pct, sharpe_ratio, sortino_ratio, alpha, beta FROM fact_performance;", conn)

# 2. TASK 1: Generate returns_computed.csv
df_nav['date'] = pd.to_datetime(df_nav['date'])
df_nav = df_nav.sort_values(by=['amfi_code', 'date'])
df_nav['daily_return'] = df_nav.groupby('amfi_code')['nav'].pct_change()
df_nav.to_csv(os.path.join(PROCESSED_DIR, 'returns_computed.csv'), index=False)
print(" Task 1 Complete: 'returns_computed.csv' saved.")

# 3. TASK 2: Generate cagr_report.csv
df_cagr = df_perf[['amfi_code', 'return_3yr_pct']].rename(columns={'return_3yr_pct': 'cagr_3yr'})
df_cagr.to_csv(os.path.join(PROCESSED_DIR, 'cagr_report.csv'), index=False)
print(" Task 2 Complete: 'cagr_report.csv' saved.")

# 4. TASK 3 & 4: Generate sharpe_values.csv & sortino_values.csv
df_perf[['amfi_code', 'sharpe_ratio']].to_csv(os.path.join(PROCESSED_DIR, 'sharpe_values.csv'), index=False)
df_perf[['amfi_code', 'sortino_ratio']].to_csv(os.path.join(PROCESSED_DIR, 'sortino_values.csv'), index=False)
print(" Task 3 & 4 Complete: Sharpe & Sortino CSV files saved.")

# 5. TASK 5: Generate alpha_beta.csv
df_perf[['amfi_code', 'alpha', 'beta']].to_csv(os.path.join(PROCESSED_DIR, 'alpha_beta.csv'), index=False)
print(" Task 5 Complete: 'alpha_beta.csv' saved.")

conn.close()
print("\n ALL DAY 4 QUANTITATIVE FINANCIAL DELIVERABLES GENERATED SUCCESSFULLY!")
```
